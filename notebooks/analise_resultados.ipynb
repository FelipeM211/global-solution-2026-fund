# Análise de Resultados
Este notebook realiza a análise de desempenho, Optimization Gap e Escala de Decisão baseada nas ODS.


```python
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregar todos os arquivos JSON da pasta data/processed/
data_dir = 'data/processed/'
data_files = [f for f in os.listdir(data_dir) if f.endswith('.json')]
datasets = {}
for f in data_files:
    with open(os.path.join(data_dir, f), 'r') as file:
        datasets[f] = json.load(file)

print('Arquivos carregados:', list(datasets.keys()))
```

## Análise de Desempenho
Abaixo, apresentamos métricas de desempenho extraídas dos dados carregados.


```python
# Exemplo de análise de desempenho - supondo que cada JSON tem uma chave 'performance'
performances = {}
for name, data in datasets.items():
    if 'performance' in data:
        performances[name] = data['performance']

# Converter para DataFrame e plotar
if performances:
    df = pd.DataFrame(performances).T
    df.plot(kind='bar', figsize=(10,6))
    plt.title('Desempenho por Arquivo')
    plt.ylabel('Valor')
    plt.xticks(rotation=45)
    plt.show()
else:
    print('Nenhum dado de desempenho encontrado.')
```

## Optimization Gap
O Optimization Gap representa a diferença entre o valor atual e o valor ótimo esperado.


```python
# Cálculo do Optimization Gap
gaps = {}
for name, data in datasets.items():
    if 'current' in data and 'optimal' in data:
        gaps[name] = (data['optimal'] - data['current']) / data['optimal'] * 100

if gaps:
    plt.figure(figsize=(8,5))
    plt.bar(gaps.keys(), gaps.values())
    plt.title('Optimization Gap (%)')
    plt.ylabel('Gap (%)')
    plt.xticks(rotation=45)
    plt.show()
else:
    print('Dados insuficientes para calcular Optimization Gap.')
```

## Escala de Decisão baseada nas ODS
A Escala de Decisão pondera contribuições para os Objetivos de Desenvolvimento Sustentável (ODS).


```python
# Exemplo de cálculo de Escala de Decisão - supondo que cada JSON tem 'ods_contributions'
scores = {}
for name, data in datasets.items():
    if 'ods_contributions' in data:
        # Soma ponderada simples (pesos iguais para cada ODS)
        contributions = data['ods_contributions']
        if isinstance(contributions, list):
            scores[name] = sum(contributions) / len(contributions)
        elif isinstance(contributions, dict):
            scores[name] = sum(contributions.values()) / len(contributions)

if scores:
    df_scores = pd.DataFrame(list(scores.items()), columns=['Arquivo', 'Score ODS'])
    df_scores.sort_values('Score ODS', ascending=False, inplace=True)
    print(df_scores)
else:
    print('Nenhum dado de contribuição ODS encontrado.')
```
